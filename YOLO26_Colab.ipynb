## Configure paths

Set `DATA_YAML` to your dataset config in Google Drive. The YAML should follow the standard Ultralytics dataset format.


```python
MODEL_NAME = 'yolo26n.pt'
DATA_YAML = '/content/drive/MyDrive/AdvReal/data.yaml'
PROJECT_DIR = '/content/drive/MyDrive/AdvReal/runs'
RUN_NAME = 'yolo26n-exp'
EPOCHS = 100
IMGSZ = 640
BATCH = 16
```

```python
from ultralytics import YOLO

model = YOLO(MODEL_NAME)
model.info()
```

## Train

This follows the official Ultralytics YOLO26 API: `model.train(data=..., epochs=..., imgsz=...)`.


```python
results = model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMGSZ,
    batch=BATCH,
    project=PROJECT_DIR,
    name=RUN_NAME,
    device=0,
)
results
```

## Validate

For end-to-end validation, YOLO26 uses the one-to-one head by default. To switch to the one-to-many head, pass `end2end=False`.


```python
metrics = model.val(data=DATA_YAML)
metrics
```

## Predict

Point `SOURCE` to an image, folder, or video in Drive.


```python
SOURCE = '/content/drive/MyDrive/AdvReal/test-images'

predictions = model.predict(
    source=SOURCE,
    save=True,
    project=PROJECT_DIR,
    name=f'{RUN_NAME}-predict',
)
predictions[0]
```

## Export

By default, YOLO26 exports the one-to-one end-to-end head. Use `end2end=False` if you need the one-to-many head.


```python
export_path = model.export(format='onnx')
export_path
```
